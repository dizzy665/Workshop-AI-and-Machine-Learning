
Question 1

Write a Python program that converts units of length, weight, and volume using functions.
---



```python
def unit_converter(value, conversion_type, direction):

    if conversion_type == "length":
        if direction == 1:
            return value * 3.28084   # meters to feet
        else:
            return value / 3.28084   # feet to meters

    elif conversion_type == "weight":
        if direction == 1:
            return value * 2.20462   # kg to pounds
        else:
            return value / 2.20462   # pounds to kg

    elif conversion_type == "volume":
        if direction == 1:
            return value * 0.264172  # liters to gallons
        else:
            return value / 0.264172  # gallons to liters


print(unit_converter(10,"length",1))
```

Question 2

Write functions to compute Sum, Average, Maximum and Minimum from a list.

```python
def find_sum(numbers):
    return sum(numbers)

def find_average(numbers):
    return sum(numbers)/len(numbers)

def find_max(numbers):
    return max(numbers)

def find_min(numbers):
    return min(numbers)


numbers = [1,2,3,4,5]

print("Sum:",find_sum(numbers))
print("Average:",find_average(numbers))
print("Max:",find_max(numbers))
print("Min:",find_min(numbers))
```

4.3 Exercise on Nested List
Question 1

Flatten a nested list.

```python
def flatten(lst):
    result=[]

    for i in lst:
        if isinstance(i,list):
            result.extend(i)
        else:
            result.append(i)

    return result


nested=[[1,2],[3,4],[5,6]]

print(flatten(nested))
```

Question 2

Access an element from a nested list using indices.

```python
def access_nested_element(lst,indices):

    element=lst

    for i in indices:
        element=element[i]

    return element


nested=[[1,2],[3,[4,5]],6]

print(access_nested_element(nested,[1,1,0]))
```

Question 3

Find the sum of elements in a nested list.

```python
def sum_nested(lst):

    total=0

    for i in lst:
        if isinstance(i,list):
            total+=sum_nested(i)
        else:
            total+=i

    return total


nested=[[1,2],[3,[4,5]],6]

print(sum_nested(nested))
```

Question 4

Remove a specific element from a nested list.

```python
def remove_element(lst,elem):

    result=[]

    for i in lst:
        if isinstance(i,list):
            result.append(remove_element(i,elem))
        else:
            if i!=elem:
                result.append(i)

    return result


nested=[[1,2],[3,[4,5]],6]

print(remove_element(nested,3))
```

Question 5

Find the maximum element in a nested list.

```python
def find_max_nested(lst):

    max_val=float('-inf')

    for i in lst:
        if isinstance(i,list):
            max_val=max(max_val,find_max_nested(i))
        else:
            max_val=max(max_val,i)

    return max_val


nested=[[1,2],[3,[4,5]],6]

print(find_max_nested(nested))
```

Question 6

Count occurrences of an element in a nested list.

```python
def count_occurrences(lst,elem):

    count=0

    for i in lst:
        if isinstance(i,list):
            count+=count_occurrences(i,elem)
        else:
            if i==elem:
                count+=1

    return count


nested=[[1,2],[2,[4,5]],2]

print(count_occurrences(nested,2))
```

Question 7

Deep flatten a nested list.

```python
def deep_flatten(lst):

    result=[]

    for i in lst:
        if isinstance(i,list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)

    return result


nested=[[1,2],[3,[4,5]],6]

print(deep_flatten(nested))
```

Question 8

Find the average of numbers in a nested list.

```python
def average_nested(lst):

    flat=deep_flatten(lst)

    return sum(flat)/len(flat)


nested=[[1,2],[3,[4,5]],6]

print(average_nested(nested))
```
